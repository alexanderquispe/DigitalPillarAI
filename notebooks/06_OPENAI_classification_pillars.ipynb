```python
import pandas as pd
from PyPDF2 import PdfReader
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import ElasticVectorSearch, Pinecone, Weaviate, FAISS
import openai
import os
openai.api_key = "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"
os.environ["OPENAI_API_KEY"] = "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"

def process_pdf(pdf_path):
    """Processes a PDF file and returns a DataFrame with the results."""

    reader = PdfReader(pdf_path)

    # read data from the file and put them into a variable called raw_text
    raw_text = ''
    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        if text:
            raw_text += text

    # We need to split the text that we read into smaller chunks so that during information retreival we don't hit the token size limits. 
    text_splitter = CharacterTextSplitter(        
        separator = "\n",
        chunk_size = 1500,
        chunk_overlap  = 300,
        length_function = len,
    )
    texts = text_splitter.split_text(raw_text)

    # Download embeddings from OpenAI
    embeddings = OpenAIEmbeddings()

    docsearch = FAISS.from_texts(texts, embeddings)

    from langchain.chains.question_answering import load_qa_chain
    from langchain.llms import OpenAI

    chain = load_qa_chain(OpenAI(), chain_type="stuff")

    # define your topic lists
    topic_lists = [infrastructure_str, platforms_str, financial_services_str, 
               businesses_str, skills_str, safeguards_str, use_cases_str]
    topic_names = ['infrastructure', 'platforms', 'financial_services', 
               'businesses', 'skills', 'safeguards', 'use_cases']

    data = {'project': os.path.basename(pdf_path).rstrip(".pdf")}  # Initialize the dictionary with the project name
    
    i = 0
    # process each list
    for topic, name in zip(topic_lists, topic_names):
        # construct the query
        query = f"Do the contents of the document mention any of the following topics: {topic}?"
        # perform the search and process the response
        docs = docsearch.similarity_search(query)
        response = chain.run(input_documents=docs, question=query)
        i = i+1
        print(i)
        print(response)
        #response_dict = json.loads(response)
        #print(str(response['usage']['total_tokens']))
        
        if 'yes' in response.lower():
            data[name] = 1
        else:
            data[name] = 0

    return pd.DataFrame(data, index=[0])


if __name__ == "__main__":
    # Get the path to the folder containing the PDF files
    pdf_folder = r"C:\Users\Alexander\Dropbox\TestSamples"

    # Process all PDF files in the folder
    pdf_files = os.listdir(pdf_folder)
    df = pd.DataFrame()
    for pdf_file in pdf_files:
        if pdf_file.endswith(".pdf"):
            df = df.append(process_pdf(os.path.join(pdf_folder, pdf_file)), ignore_index=True)

    # Print the DataFrame
    print(df)
```

```python
from PyPDF2 import PdfReader
import openai
# Load your OpenAI API key
openai.api_key = "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"

# Define the topics you're interested in
topics = [
    "Broadband infrastructure",
    "Internet infrastructure",
    "Network infrastructure",
    "Connectivity infrastructure",
    "Fiber infrastructure",
    "Broadband connectivity",
    "Broadband",
    "digital inclusion",
    "network connectivity",
    "Telecommunications",
    "Mobile operator",
    "Telecom operator",
    "Network operator",
    "Submarine network",
    "Submarine cable",
    "Satellite",
    "Cable landing",
    "Internet cable",
    "Submarine fiber",
    "Submarine connectivity",
    "virtual landing station",
    "cross-border connectivity",
    "international gateway",
    "Backbone network",
    "national backbone",
    "fiber backbone",
    "microwave backbone",
    "backhaul network",
    "Local access network",
    "Fiber-to-the-home",
    "FTTH",
    "WIFI",
    "public access point",
    "community center",
    "local loop network",
    "universal service fund (USF)",
    "universal access fund (UAF)",
    "USAF",
    "broadband access",
    "Network coverage",
    "mobile coverage",
    "digital gender gap",
    "gender digital divide",
    "colocation data center",
    "cloud storage",
    "cloud platforms",
    "data center",
    "cloud computing",
    "telecom SOE",
    "telecom state-owned enterprise",
    "telecom market liberalization",
    "open to competition",
    "telecom license",
    "mobile license",
    "IXP",
    "ICT regulator",
    "telecom regulator",
    "interconnection",
    "Infrastructure sharing",
    "telecom license",
    "spectrum management",
    "telecom tax",
    "digital tax",
    "national roaming"
]

def extract_text_from_pdf(file_path):
    pdf = PdfReader(file_path)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text

def check_topics_in_text(text, topics, context_window=500):
    results = {}
    for topic in topics:
        # Search for the topic in the text
        topic_position = text.lower().find(topic.lower())
        if topic_position == -1:
            # The topic was not found
            results[topic] = False
        else:
            # Extract the context window around the topic
            start = max(0, topic_position - context_window)
            end = min(len(text), topic_position + len(topic) + context_window)
            context = text[start:end]

            # Ask the model whether the topic is discussed in the context
            prompt = f"The following text is about {topic}: {context}"
            response = openai.Completion.create(engine="text-davinci-003", prompt=prompt, max_tokens=10)
            result = response.choices[0].text.strip()
            results[topic] = result.lower() in ['yes', 'true', 'probably', 'likely']
    return results

pdf_path = "C:/Users/Alexander/Dropbox/TestSamples/P160766_PAD.pdf"
pdf_text = extract_text_from_pdf(pdf_path)
topic_results = check_topics_in_text(pdf_text, topics)

for topic, is_discussed in topic_results.items():
    print(f"Is {topic} discussed? {is_discussed}")
```

```python
import tensorflow_hub as hub
import numpy as np
from PyPDF2 import PdfReader

# Load Universal Sentence Encoder
model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

def extract_text_from_pdf(file_path):
    pdf = PdfReader(file_path)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text

def check_topics_in_text(text, topics):
    results = {}
    
    # Create an embedding for the text
    text_embedding = model([text])[0]
    
    for topic in topics:
        # Create an embedding for the topic
        topic_embedding = model([topic])[0]
        
        # Calculate the cosine similarity between the text and the topic
        cosine_similarity = np.inner(text_embedding, topic_embedding)
        
        # If the cosine similarity is above a certain threshold, we can say the topic is discussed
        results[topic] = cosine_similarity > 0.5

    return results

# Load your topics
#topics = ["Your topics here"]
#pdf_path = 'path_to_your_pdf.pdf'
pdf_text = extract_text_from_pdf(pdf_path)
topic_results = check_topics_in_text(pdf_text, topics)

for topic, is_discussed in topic_results.items():
    print(f"Is {topic} discussed? {is_discussed}")
```

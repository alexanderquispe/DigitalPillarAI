```python
import os
import pandas as pd
import json

# Definimos la ruta donde se encuentran los archivos JSON
ruta_json = './output_json/'

# Inicializamos una lista para almacenar los DataFrames de cada archivo JSON
dataframes = []

# Iteramos sobre cada archivo en la ruta
for archivo in os.listdir(ruta_json):
    # Verificamos que sea un archivo JSON
    if archivo.endswith('.json'):
        # Leemos el archivo JSON y convertimos los datos en un diccionario de Python
        with open(os.path.join(ruta_json, archivo)) as f:
            data = json.load(f)
            docs = data['documents']
            docs.pop('facets', None)  # Ignoramos la propiedad "facets" dentro de "documents"
        
        # Creamos un DataFrame con los datos del archivo JSON y lo agregamos a la lista
        df = pd.DataFrame.from_dict(docs, orient='index')
        dataframes.append(df)

# Combinamos todos los DataFrames en uno solo
df_combinado = pd.concat(dataframes)

# Reiniciamos los índices del DataFrame combinado
# df_combinado.reset_index(drop=True, inplace=True)

# Mostramos el DataFrame combinado
# print(df_combinado)
# Guardar el DataFrame combinado como un archivo Excel
ruta_salida_excel = 'data.csv'
df_combinado.to_csv(ruta_salida_excel, index=False)
```
